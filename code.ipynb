```python
import cv2
import numpy as np

def camera_access():
    
    cap = cv2.VideoCapture(0)
    
    if not cap.isOpened():  
        print("Error: Could not access the camera.")
        return
    
    print("Press 'q' to quit")
    
    try:
        while True:
            ret, frame = cap.read()
            
            if not ret:
                print("Error: Can't receive frame")
                break
                
            cv2.imshow('Camera Feed', frame)
            
            key = cv2.waitKey(1)
            if key & 0xFF == ord('q'):
                print("Quitting camera feed...")
                break
                
    finally:
        cap.release()
        cv2.destroyAllWindows()

camera_access()
```

```python

import cv2
import numpy as np

def run_camera_sentiment(camera_index=0, smile_threshold=1):
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")

    cap = cv2.VideoCapture(camera_index)
    if not cap.isOpened():
        print("Error: could not open camera")
        return

    print("Camera started. Press 'q' to quit.")

    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Frame grab failed")
                break

            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(60, 60))

            for (x, y, w, h) in faces:
                face_roi_gray = gray[y:y+h, x:x+w]
                face_roi_color = frame[y:y+h, x:x+w]

                smiles = smile_cascade.detectMultiScale(face_roi_gray, scaleFactor=1.7, minNeighbors=20)
                sentiment = "Neutral"
                color = (255, 255, 0)  
                if len(smiles) >= smile_threshold:
                    sentiment = "Positive"
                    color = (0, 255, 0) 
                    
                cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
                cv2.putText(frame, sentiment, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

                for (sx, sy, sw, sh) in smiles:
                    cv2.rectangle(face_roi_color, (sx, sy), (sx+sw, sy+sh), (0, 200, 200), 1)

            cv2.imshow("Real-time Sentiment (face->emotion)", frame)

            key = cv2.waitKey(1)
            if key != -1 and (key & 0xFF) == ord('q'):
                break

    finally:
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    run_camera_sentiment()
```

```python

import cv2
import numpy as np
from ultralytics import YOLO
import time
import math

model = YOLO("yolov8n.pt")

next_object_id = 0
prev_centroids = {}  
directions = {}       
last_seen = {}        
frame_idx = 0


FOCAL_LENGTH = 800.0  
KNOWN_HEIGHTS = {"person": 1.70} 
def estimate_distance_m(label, bbox_height_px, focal_length=FOCAL_LENGTH):
    if bbox_height_px <= 0:
        return None
    h_m = KNOWN_HEIGHTS.get(label, None)
    if h_m is None:
        return None
    
    return (h_m * focal_length) / bbox_height_px

def match_detections_to_prev(centroids, max_dist=80):
    """
    centroids: list of (cx, cy)
    returns list of assigned ids (same length as centroids)
    """
    global next_object_id, prev_centroids, last_seen, frame_idx
    assigned_ids = [-1] * len(centroids)
    used_prev_ids = set()
    
    
    for i, c in enumerate(centroids):
        best_id = None
        best_dist = None
        for oid, pc in prev_centroids.items():
            if oid in used_prev_ids:
                continue
            d = math.hypot(c[0] - pc[0], c[1] - pc[1])
            if d <= max_dist and (best_dist is None or d < best_dist):
                best_dist = d
                best_id = oid
        if best_id is not None:
            assigned_ids[i] = best_id
            used_prev_ids.add(best_id)
            prev_centroids[best_id] = c
            last_seen[best_id] = frame_idx
        else:
            
            oid = next_object_id
            next_object_id += 1
            prev_centroids[oid] = c
            last_seen[oid] = frame_idx
            directions[oid] = "Static"
            assigned_ids[i] = oid
            
            
    remove = [oid for oid, seen in last_seen.items() if frame_idx - seen > 100]
    for oid in remove:
        prev_centroids.pop(oid, None)
        last_seen.pop(oid, None)
        directions.pop(oid, None)
    return assigned_ids

def run_camera_yolo(camera_index=0, conf=0.35):
    global frame_idx, prev_centroids, directions
    cap = cv2.VideoCapture(camera_index)
    if not cap.isOpened():
        print("Error: cannot open camera")
        return

    prev_time = time.time()
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            frame_idx += 1

            results = model(frame, conf=conf)[0]

            detections = []
            centroids = []
            labels = []
            scores = []
            
            if results.boxes is not None and len(results.boxes) > 0:
                for box in results.boxes:
                    xyxy = box.xyxy[0].cpu().numpy().astype(int)
                    x1, y1, x2, y2 = xyxy
                    cls = int(box.cls[0].cpu().numpy())
                    score = float(box.conf[0].cpu().numpy())
                    label = model.names[cls] if hasattr(model, "names") else str(cls)
                    detections.append((x1, y1, x2, y2))
                    cx = int((x1 + x2) / 2)
                    cy = int((y1 + y2) / 2)
                    centroids.append((cx, cy))
                    labels.append(label)
                    scores.append(score)

            ids = match_detections_to_prev(centroids)

            for idx, det in enumerate(detections):
                x1, y1, x2, y2 = det
                cid = ids[idx]
                label = labels[idx]
                score = scores[idx]
                cx, cy = centroids[idx]

                prev_c = prev_centroids.get(cid, (cx, cy))
                dx = cx - prev_c[0]
                dir_text = directions.get(cid, "Static")
                
                if abs(dx) > 5:
                    new_dir = "L->R" if dx > 0 else "R->L"
                    directions[cid] = new_dir
                    dir_text = new_dir

                bbox_h = max(1, y2 - y1)
                dist_m = estimate_distance_m(label, bbox_h)
                dist_text = f"{dist_m:.2f} m" if dist_m is not None else "N/A"

                color = (0, 255, 0) if label in ("person", "human") or label == "person" else (255, 165, 0)
                cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

                txt1 = f"{label} {score:.2f}"
                txt2 = f"ID:{cid} {dir_text} {dist_text}"
                cv2.putText(frame, txt1, (x1, y1 - 18), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
                cv2.putText(frame, txt2, (x1, y1 - 2), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

                cv2.circle(frame, (cx, cy), 3, color, -1)

            cur_time = time.time()
            fps = 1.0 / (cur_time - prev_time) if cur_time != prev_time else 0.0
            prev_time = cur_time
            cv2.putText(frame, f"FPS: {fps:.1f}", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)

            cv2.imshow("YOLOv8 Real-time Object Detection (dir + dist)", frame)
            key = cv2.waitKey(1) & 0xFF
            if key == ord('q') or key == 27:
                break
    finally:
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    run_camera_yolo()
```

```python
def run_camera_yolo(camera_index=0, conf=0.35, flip_horizontal=True):
    global frame_idx, prev_centroids, directions
    cap = cv2.VideoCapture(camera_index)
    if not cap.isOpened():
        print("Error: cannot open camera")
        return

    prev_time = time.time()
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
                
            if flip_horizontal:
                frame = cv2.flip(frame, 1)  
                
            frame_idx += 1
            
            results = model(frame, conf=conf)[0]
            
            detections = []
            centroids = []
            labels = []
            scores = []
            
            if results.boxes is not None and len(results.boxes) > 0:
                for box in results.boxes:
                    xyxy = box.xyxy[0].cpu().numpy().astype(int)
                    x1, y1, x2, y2 = xyxy
                    cls = int(box.cls[0].cpu().numpy())
                    score = float(box.conf[0].cpu().numpy())
                    label = model.names[cls] if hasattr(model, "names") else str(cls)
                    detections.append((x1, y1, x2, y2))
                    cx = int((x1 + x2) / 2)
                    cy = int((y1 + y2) / 2)
                    centroids.append((cx, cy))
                    labels.append(label)
                    scores.append(score)

            ids = match_detections_to_prev(centroids)

            for idx, det in enumerate(detections):
                x1, y1, x2, y2 = det
                cid = ids[idx]
                label = labels[idx]
                score = scores[idx]
                cx, cy = centroids[idx]

                prev_c = prev_centroids.get(cid, (cx, cy))
                dx = cx - prev_c[0]
                dir_text = directions.get(cid, "Static")
                
                
                if abs(dx) > 5:
                    if flip_horizontal:
                        new_dir = "L->R" if dx < 0 else "R->L"  # Reversed for mirrored view
                    else:
                        new_dir = "L->R" if dx > 0 else "R->L"
                    directions[cid] = new_dir
                    dir_text = new_dir

                bbox_h = max(1, y2 - y1)
                dist_m = estimate_distance_m(label, bbox_h)
                dist_text = f"{dist_m:.2f} m" if dist_m is not None else "N/A"

                color = (0, 255, 0) if label in ("person", "human") or label == "person" else (255, 165, 0)
                cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

                txt1 = f"{label} {score:.2f}"
                txt2 = f"ID:{cid} {dir_text} {dist_text}"
                cv2.putText(frame, txt1, (x1, y1 - 18), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
                cv2.putText(frame, txt2, (x1, y1 - 2), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

                cv2.circle(frame, (cx, cy), 3, color, -1)

            cur_time = time.time()
            fps = 1.0 / (cur_time - prev_time) if cur_time != prev_time else 0.0
            prev_time = cur_time
            cv2.putText(frame, f"FPS: {fps:.1f}", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)

            cv2.imshow("YOLOv8 Real-time Object Detection (dir + dist)", frame)
            key = cv2.waitKey(1) & 0xFF
            if key == ord('q') or key == 27:
                break
    finally:
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    run_camera_yolo(flip_horizontal=True)
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import time

class ObjectDetectorTD:
    def __init__(self):
        self.model = YOLO("yolov8n.pt")
        self.prev_time = time.time()
        
    def process_frame(self, frame_in):
        try:
            frame = frame_in.copy()
            results = self.model(frame, conf=0.35)[0]
            
            for box in results.boxes:
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
                cls = int(box.cls[0].cpu().numpy())
                conf = float(box.conf[0].cpu().numpy())
                label = self.model.names[cls]
                
                color = (0, 255, 0) if label == "person" else (255, 165, 0)
                cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
                
                text = f"{label} {conf:.2f}"
                cv2.putText(frame, text, (x1, y1-10), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
                
            cur_time = time.time()
            fps = 1.0 / (cur_time - self.prev_time)
            self.prev_time = cur_time
            cv2.putText(frame, f"FPS: {fps:.1f}", (10, 25),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)
                
            return frame
            
        except Exception as e:
            print(f"Error processing frame: {e}")
            return frame_in

class ObjectDetectionTOP:
    def __init__(self, ownerComp):
        self.ownerComp = ownerComp
        self.detector = ObjectDetectorTD()
        
    def onCook(self, dat):
        try:
            op = self.ownerComp.inputs[0]
            frame = op.numpyArray(delayed=True)
            processed = self.detector.process_frame(frame)
            if processed is not None:
                dat.copyNumpyArray(processed)
        except Exception as e:
            print(f"Error in TOP cook: {e}")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import time

class CameraDetectorTD:
    def __init__(self):
        self.model = YOLO("yolov8n.pt")
        self.prev_time = time.time()
        self.cap = cv2.VideoCapture(0)
        self.frame_count = 0
        
    def __del__(self):
        self.cap.release()
        
    def get_frame(self):
        try:
            ret, frame = self.cap.read()
            if not ret:
                return None
                
            frame = cv2.flip(frame, 1)
            results = self.model(frame, conf=0.35)[0]
            
            for box in results.boxes:
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
                cls = int(box.cls[0].cpu().numpy())
                conf = float(box.conf[0].cpu().numpy())
                label = self.model.names[cls]
                
                color = (0, 255, 0) if label == "person" else (255, 165, 0)
                cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
                
                text = f"{label} {conf:.2f}"
                cv2.putText(frame, text, (x1, y1-10), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
                
            cur_time = time.time()
            fps = 1.0 / (cur_time - self.prev_time)
            self.prev_time = cur_time
            cv2.putText(frame, f"FPS: {fps:.1f}", (10, 25),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)
            
            self.frame_count += 1
            return frame
            
        except Exception as e:
            print(f"Error capturing frame: {e}")
            return None

class CameraDetectionTOP:
    def __init__(self, ownerComp):
        self.ownerComp = ownerComp
        self.detector = CameraDetectorTD()
        
    def onCook(self, dat):
        try:
            frame = self.detector.get_frame()
            if frame is not None:
                dat.copyNumpyArray(frame)
        except Exception as e:
            print(f"Error in TOP cook: {e}")
            
    def onPulse(self, par):
        if par.name == "Reset":
            del self.detector
            self.detector = CameraDetectorTD()

def onCreateParams(params):
    page = params.appendPage('Camera')
    
    p = page.appendFloat('Confidence', label='Detection Confidence')
    p.default = 0.35
    p.min = 0
    p.max = 1
    
    p = page.appendPulse('Reset', label='Reset Camera')
    
```
